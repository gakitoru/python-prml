```python
%matplotlib inline 
import numpy as np
import math
import matplotlib.pyplot as plt
#from scipy.stats import norm

N = 10
X = np.linspace(0, 1, N)
T = [math.sin(2*math.pi*x) + np.random.normal(loc=0, scale=0.2) for x in X]

print(X)
print(T)
plt.scatter(X, T)

X_ref = np.linspace(0, 1, 1000)
T_ref = [math.sin(2*math.pi*x) for x in X_ref]
T_ref_noise = [t + np.random.normal(loc=0, scale=0.2) for t in T_ref]

plt.plot(X_ref, T_ref, c='g')
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X.reshape(len(X), 1), T)
T_linear_one = lr.predict(X_ref.reshape(len(X_ref), 1))

plt.scatter(X, T)
plt.plot(X_ref, T_ref, c='g')
plt.plot(X_ref.T, T_linear_one)
plt.show()
```

```python
dims = np.arange(0, 10)

print(X)
X_train_dim_list = []
X_ref_dim_list = []
for dim in dims:
    print("dim = ", dim)
    X_train_dim = []
    for x in X:
        x_train_dim = []
        for i in range(0, dim + 1):
            x_train_dim.append(x**i)
        X_train_dim.append(x_train_dim)
        
    X_ref_dim = []
    for x in X_ref:
        x_ref_dim = []
        for i in range(0, dim + 1):
            x_ref_dim.append(x**i)
        X_ref_dim.append(x_ref_dim)
    X_train_dim_list.append(X_train_dim)
    X_ref_dim_list.append(X_ref_dim)
```

```python
T_ref
```

```python
Erms_train = []
Erms_test = []

for dim in dims:
    print("M = ", + dim)
    X_train_dim = np.array(X_train_dim_list[dim])
    X_ref_dim = np.array(X_ref_dim_list[dim])

    lr = LinearRegression()
    lr.fit(X_train_dim.reshape(len(X_train_dim), -1), T)
    
    #print(X_ref_dim)
    T_ref_dim = lr.predict(X_ref_dim)

    
    T_train_dim = lr.predict(X_train_dim)
    Erms = 0
    print(T_train_dim)
    print(T)
    print(math.sqrt(2*sum((T_train_dim - T)**2)/len(X_train_dim)))
    Erms_train.append(math.sqrt(2*sum((T_train_dim - T)**2)/len(X_train_dim)))
    Erms_test.append(math.sqrt(2*sum((T_ref_dim - T_ref_noise)**2)/len(X_ref_dim)))
    
    #math.sqrt(2*((T_train_dim - T)**2)/len(X_train_dim))
    
    plt.scatter(X, T)
    plt.plot(X_ref, T_ref_dim)
    plt.plot(X_ref, T_ref, c='g')
    plt.show()
    
plt.scatter(dims, Erms_test)
plt.scatter(dims, Erms_train)
plt.plot(dims, Erms_test)
plt.plot(dims, Erms_train)
plt.show()

    
```

```python
from sklearn.linear_model import Ridge

Erms_train.clear()
Erms_test.clear()

for dim in dims:
    print("M = ", + dim)
    X_train_dim = np.array(X_train_dim_list[dim])
    X_ref_dim = np.array(X_ref_dim_list[dim])

    rd = Ridge(alpha=100000)
    rd.fit(X_train_dim.reshape(len(X_train_dim), -1), T)
    
    #print(X_ref_dim)
    T_ref_dim = rd.predict(X_ref_dim)

    
    T_train_dim = rd.predict(X_train_dim)
    Erms = 0
    print(T_train_dim)
    print(T)
    print(math.sqrt(2*sum((T_train_dim - T)**2)/len(X_train_dim)))
    Erms_train.append(math.sqrt(2*sum((T_train_dim - T)**2)/len(X_train_dim)))
    Erms_test.append(math.sqrt(2*sum((T_ref_dim - T_ref_noise)**2)/len(X_ref_dim)))
    
    #math.sqrt(2*((T_train_dim - T)**2)/len(X_train_dim))
    
    plt.scatter(X, T)
    plt.plot(X_ref, T_ref_dim)
    plt.plot(X_ref, T_ref, c='g')
    plt.show()
    
plt.scatter(dims, Erms_test)
plt.scatter(dims, Erms_train)
plt.plot(dims, Erms_test)
plt.plot(dims, Erms_train)
plt.show()
```
