```python
%matplotlib inline 
import numpy as np
import math
import matplotlib.pyplot as plt
#from scipy.stats import norm

X = np.linspace(0, 1, 10)
T = [math.sin(2*math.pi*x) + np.random.normal(loc=0, scale=0.2) for x in X]

print(X)
print(T)
plt.scatter(X, T)

X_ref = np.linspace(0, 1, 1000)
T_ref = [math.sin(2*math.pi*x) for x in X_ref]

plt.plot(X_ref, T_ref, c='g')
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X.reshape(len(X), 1), T)
T_linear_one = lr.predict(X_ref.reshape(len(X_ref), 1))

plt.scatter(X, T)
plt.plot(X_ref, T_ref, c='g')
plt.plot(X_ref.T, T_linear_one)
plt.show()
```

```python
dims = np.arange(0, 10)

print(X)
X_train_list = []
X_test_list = []
for dim in dims:
    print("dim = ", dim)
    X_train = []
    for x in X:
        x_train = []
        for i in range(0, dim + 1):
            x_train.append(x**i)
        X_train.append(X_train)
        
    X_test = []
    for x in X_ref:
        x_test = []
        for i in range(0, dim + 1):
            x_test.append(x**i)
        X_test.append(X_test)
    X_train_list.append(X_train)
    X_test_list.append(X_test)
```

```python
for dim in dims:
    X_train = np.array(X_train_list[dim])
    X_test = np.array(X_test_list[dim])

    lr = LinearRegression()
    lr.fit(X_train.reshape(len(X_train), -1), T)
    
    T_test = lr.predict(X_test)

    plt.scatter(X, T)
    plt.plot(X_ref, T_test)
    plt.plot(X_ref, T_ref, c='g')
    plt.show()
```
